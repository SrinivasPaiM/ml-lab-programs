```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Generate 100 random values between 0 and 1
X = np.random.rand(100).reshape(-1, 1)

# Step 2: Label first 50 points
y_train = ["Class1" if x <= 0.5 else "Class2" for x in X[:50, 0]]
X_train = X[:50]
X_test = X[50:]

# Step 3: Use sklearn's KNeighborsClassifier
k_values = [1, 2, 3, 4, 5, 20, 30]
for k in k_values:
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    
    print(f"\n--- Results for k = {k} ---")
    for i, (x_val, pred) in enumerate(zip(X_test, y_pred), start=51):
        print(f"x{i} (value: {x_val[0]:.4f}) → {pred}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Generate 100 random values in [0,1]
X = np.random.rand(100).reshape(-1, 1)

# Step 2: Label first 50 points: if xi ≤ 0.5 -> Class 0, else -> Class 1
y_train = np.array([0 if x <= 0.5 else 1 for x in X[:50]])

# Step 3: Define test data
X_train = X[:50]
X_test = X[50:]

# Step 4: Try different values of k
k_values = [1, 2, 3, 4, 5, 20, 30]

for k in k_values:
    # Train k-NN classifier
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Predict for test data
    y_pred = knn.predict(X_test)
    
    print(f"\n--- k = {k} ---")
    print("Predicted classes for X[51 to 100]:")
    print(y_pred)
    
    # Plot results
    plt.figure(figsize=(8, 3))
    
    # Plot training points
    plt.scatter(X_train, [0.1]*len(X_train), c=y_train, cmap='bwr', label='Training (0.1)', marker='o')
    
    # Plot test points with predicted classes
    plt.scatter(X_test, [0.9]*len(X_test), c=y_pred, cmap='bwr', label='Test (0.9)', marker='x')
    
    plt.title(f'k-NN Classification for k = {k}')
    plt.xlabel('x value')
    plt.yticks([0.1, 0.9], ['Train', 'Test'])
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
